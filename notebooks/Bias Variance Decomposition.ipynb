```python
from sklearn import metrics
import pandas as pd
import numpy as np
```

```python
# Let's say we are getting 100 samples from a population
ts_samples = []
ps_samples = []

bs = []
vs = []

mses = []
maes = []
for s_idx in range(200):
    # Each sample contains 1000 data points
    # Make 1000 targets from a normal distribution
    ts = np.random.randn(1000)
    # Make 1000 random prediction from a normal distribution
    ps = np.random.randn(1000)
    
    mses.append(metrics.mean_squared_error(ts, ps))
    maes.append(metrics.mean_absolute_error(ts, ps))
    
    bias = (ps - ts).mean()
    var = ((ps.mean() - ps) ** 2).mean()
    bs.append(bias)
    vs.append(var)
```

```python
report = pd.DataFrame({
    "bias": bs,
    "var": vs,
    "b^2 + var": np.array(bs) ** 2 + var,
    "mse": mses,
    "maes": maes,
})
report
```

```python
report.plot()
```
